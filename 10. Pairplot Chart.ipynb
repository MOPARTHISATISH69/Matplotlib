PairPlot - A simple pait plot is used to understand the relationship or pattern between various numerical variables in our dataset. We plot multiple scatter plot between various numerical variables in a single plot.

```python
# Importing libraries

import pandas as pd
import matplotlib.pyplot as plt

from pydataset import data

#using inbuilt data set in pydataset named "IRIS" data set.

I = data("iris")
I.head()
```

```python
# Let's remove the categorical column "species"

I_F = I.drop(columns = ['Species'])
I_F.head()
```

```python
# I want to know the relationship all these columns .. relationship between sepal.lenght and sepal.width 
# relationship between sepal.width  and petal.length .. relationship between petal.Length and petal.width .. so on
```

```python
# Create a pair plot matrix from the data frame

pd.plotting.scatter_matrix(I_F,figsize=(8,8),marker='o')
```
