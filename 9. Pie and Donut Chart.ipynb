Pie Chart <--> A pie chart displays data, information , and statistical in an easy-to-read 'pie slice' format with varying slice sizes telling you how much of one data element exists. The bigger the slice , the more of that particular data was gathered.

When items are presented on a pie chart, you can eaisly see which items is the most popular and which is the least popular.

A pie chart tells the composition of a categorical variable based on a numerical value

                source :: study.com

```python
# Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data

# Using inbuild dataset in pydataset named as "mpg" data set

P = data('mpg')
P.head()
```

I would like to plot percentage of cars for each manufacturer in my data set. Here manufacturer is the categorical column and the total cars for each manufacturer is the numerical column. 

```python
# Calculating the total cars for each manufacturer

P_M = P.groupby(['manufacturer']).size().reset_index()
P_M.head()
```

```python
#re-naming the column as total cars
P_M = P_M.rename(columns = {0:"Total Cars"})
P_M.head()
```

```python
# Plotting the pie chart

# We will use output to indicate percentage labels on pie chart

plt.rcParams['figure.figsize'] = [8,8]
plt.pie(P_M['Total Cars'],labels=P_M['manufacturer'],autopct='%.2f%%')
plt.show()

#'%.2f%%' --> first and last % represent that output shoule be in percentange. 
# .2f    --> upto 2 decimal values 
# .2f%   --> means , upto 2 decimal values with % symbol at end.
```

```python
## Converting the same pie chart into donut chart
## We can just change pie chart to donut chart by drwaing a circle with white colot at origin.

plt.rcParams['figure.figsize'] = [8,8]

plt.pie(P_M['Total Cars'],labels=P_M['manufacturer'],autopct='%.2f%%',pctdistance=.8)

# draw a circle centered at (0,0)

Center_Circle = plt.Circle((0,0),0.60,fc="white")

fig = plt.gcf()

fig.gca().add_artist(Center_Circle)
plt.show()
```

Default radius of pie is 1, we are drawing an circle with radius 0.6
